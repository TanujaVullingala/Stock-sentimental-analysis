```python
import pandas as pd
```

```python
dt= pd.read_csv('headlines.csv',encoding='ISO-8859-1')
```

```python
dt.head()
```

```python
train = dt[dt['Date']<'20150101']
test = dt[dt['Date']>'20141231']
```

```python
d = train.iloc[:,2:27]
d.replace("[^a-zA-Z]"," ",regex=True,inplace=True)

original=[i for i in range(25)]
newindex = [str(i) for i in original]
d.columns = newindex
d.head()
```

```python
for i in newindex:
    d[i] = d[i].str.lower()
d.head()    
```

```python
summed =[]
for i in range(0,len(d.index)):
    summed.append(' '.join(str(x) for x in d.iloc[i,0:25]))
summed[0]    
```

```python
import sklearn
```

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
do_vector= CountVectorizer(ngram_range=(2,2))
newtrain = do_vector.fit_transform(summed)
newtrain[0]
```

```python
rc_model = RandomForestClassifier(n_estimators=200,criterion = 'entropy')
rc_model.fit(newtrain,train['Label'])
```

```python
tester=[]
for i in range(0,len(test.index)):
    tester.append(' '.join(str(x) for x in test.iloc[i,2:27]))
newtester = do_vector.transform(tester)
results = rc_model.predict(newtester)
```

```python
results
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
table = confusion_matrix(test['Label'],results)
print(table)
score = accuracy_score(test['Label'],results)
print(score)
report = classification_report(test['Label'],results)
print(report)
```

```python
test.loc[:, 'Sentiment'] = results
```

```python
import yfinance as yf
```

```python
ticker = 'META'
start_date = '2000-03-01'
end_date = '2016-01-07'
stock_data = yf.download(ticker, start=start_date, end=end_date)
```

```python
stock_data.head()
```

```python
testing_new = test.copy()
```

```python
testing_new.loc[:, 'EventDate'] = pd.to_datetime(testing_new['Date'])
```

```python
stock_data.reset_index(inplace=True)
```

```python
print(stock_data['Date'].dtype)
print(testing_new['EventDate'].dtype)
```

```python

data = pd.merge(stock_data, testing_new[['EventDate', 'Sentiment']], left_on='Date', right_on='EventDate', how='left')
```

Generate Buy/Sell Signals and Simulate Trading

```python
data['signal'] = 0
data.loc[data['Sentiment'] == 1, 'signal'] = 1  # Buy signal
data.loc[data['Sentiment'] == 0, 'signal'] = -1  # Sell signal
```

```python
initial_cash = 10000
cash = initial_cash
stock_owned = 0
portfolio_value = []
```

```python
for i in range(len(data)):
    if data['signal'][i] == 1 and cash >= data['Close'][i]:
        stock_owned = cash // data['Close'][i]
        cash -= stock_owned * data['Close'][i]
    elif data['signal'][i] == -1 and stock_owned > 0:
        cash += stock_owned * data['Close'][i]
        stock_owned = 0
    portfolio_value.append(cash + stock_owned * data['Close'][i])

data['portfolio_value'] = portfolio_value
```

Plot and Analyze Results

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Stock Price')
plt.plot(data[data['signal'] == 1]['Date'], data['Close'][data['signal'] == 1], '^', markersize=10, color='g', label='Buy Signal')
plt.plot(data[data['signal'] == -1]['Date'], data['Close'][data['signal'] == -1], 'v', markersize=10, color='r', label='Sell Signal')
plt.plot(data['Date'], data['portfolio_value'], label='Portfolio Value', linestyle='--')
plt.legend()
plt.title(f'{ticker} Stock Price and Trading Signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
final_portfolio_value = portfolio_value[-1]
total_return = (final_portfolio_value - initial_cash) / initial_cash * 100
print(f'Initial cash: ${initial_cash}')
print(f'Final portfolio value: ${final_portfolio_value:.2f}')
print(f'Total return: {total_return:.2f}%')
```
